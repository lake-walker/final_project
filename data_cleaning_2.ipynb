```python
# import dependencies 
import pandas as pd
```

```python
# Bring in all the new csv files to be converted to pd dataframes
alcohol = pd.read_csv('data/total-alcohol-consumption-per-capita-litres-of-pure-alcohol.csv')
doctors = pd.read_csv('data/doctors.csv')
health_spending = pd.read_csv('data/health_spending.csv')
debt = pd.read_csv('data/debt.csv')
# social_support = pd.read_csv('data/lack_of_social_support.csv')
# obese = pd.read_csv('data/obese_rate.csv')
# years_lost = pd.read_csv('data/potential_years_life_lost.csv')
```

```python
debt.head()
```

```python
# Drop all the unneededs columns from all the dataframes
# alcohol = alcohol.drop(columns={'SUBJECT','MEASURE','FREQUENCY','Flag Codes','INDICATOR','TIME'})
doctors = doctors.drop(columns={'Indicator Name'})
health_spending = health_spending.drop(columns={'Indicator Name'})
# # social_support = social_support.drop(columns={'SUBJECT','MEASURE','FREQUENCY','Flag Codes','INDICATOR','TIME'})
# obese = obese.drop(columns={'SUBJECT','MEASURE','FREQUENCY','Flag Codes','INDICATOR','TIME'})
# years_lost = years_lost.drop(columns={'SUBJECT','MEASURE','FREQUENCY','Flag Codes','INDICATOR','TIME'})
# Rename the value column
alcohol = alcohol.rename(columns={'Entity':'country','Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)':'alcohol_consumption_percapita'})
doctors = doctors.rename(columns={'Country Name':'country','2015':'physicians_per_1k_2015','2016':'physicians_per_1k_2016','2017':'physicians_per_1k_2017'})
health_spending = health_spending.rename(columns={'Country Name':'country','2015':'health_spending_percapita_2015','2016':'health_spending_percapita_2016','2017':'health_spending_percapita_2017','2018':'health_spending_percapita_2018'})
debt = debt.rename(columns={'Private debt, loans and debt securities (Percent of GDP)':'country','2015':'private_debt_2015','2016':'private_debt_2016'})
# social_support = social_support.rename(columns={'Value':'social_support'})
# obese = obese.rename(columns={'Value':'obese'})
# years_lost = years_lost.rename(columns={'Value':'years_lost'})
```

```python
# combine the data frames on time and country
combined_df = alcohol.merge(doctors,  how='left', on='country')
combined_df = combined_df.merge(health_spending,  how='left', on='country')
combined_df = combined_df.merge(debt,  how='left', on='country')
# combined_df = combined_df.merge(obese,  how='inner', on='country')
# combined_df = combined_df.merge(years_lost,  how='inner', on='country')
combined_df[combined_df['country']=='Canada']
```

```python
# change the columns to lowercase for better querrying 
combined_df = combined_df.rename(columns=str.lower)
combined_df
```

```python
droped_df = combined_df.dropna()
droped_df.shape
```

```python
# add in the completed dataframe from project 2 to add the new data to
full_df = pd.read_csv('data/country_data.csv')
full_df = full_df.merge(combined_df, how='right', on='country')
```

```python
full_df.columns
```

```python
# determine which columns to keep and which to drop
clean_df = full_df.drop(columns={'sex','standard_error','economy_gdp_per_capita_2015', 'family_2015',\
                                 'health_life_expectancy_2015', 'freedom_2015','trust_government_corruption_2015',\
                                 'generosity_2015','economy_gdp_per_capita_2016', 'family_2016','health_life_expectancy_2016',\
                                 'freedom_2016','trust_government_corruption_2016', 'generosity_2016','dystopia_residual_2016',\
                                'economy_gdp_2016', 'economy_gdp_2015','life_expectancy_at_birth','expected_years_of_education',\
                                 'mean_years_of_education','gross_national_income_gni_per_capita',\
                                 'gni_per_capita_rank_minus_hdi_rank','physicians_per_1k_2016',\
                                 'physicians_per_1k_2017','health_spending_percapita_2016','health_spending_percapita_2017',\
                                 'health_spending_percapita_2018','sui_per_100k_2016','sui_per_100k_2010', 'sui_per_100k_2000',\
                                'dystopia_residual_2015','happiness_rank_2016', 'happiness_score_2016','total_suicides_2016',
                                'total_population', 'total_suicides_2015', 'code'})
clean_df.columns
```

```python
new_country_data = clean_df.drop(columns={'physicians_per_1k_2015', 'private_debt_2016'})
new_country_data
```

```python
new_country_data = new_country_data.dropna()
new_country_data = new_country_data.reset_index()
new_country_data = new_country_data.drop(columns={'index'})
new_country_data.shape
```

```python
# change the deby datatypes to numeric to add to region dataframe
new_country_data['private_debt_2015'] = pd.to_numeric(new_country_data['private_debt_2015'],errors='coerce')
new_country_data.dtypes
```

```python
# create region dataset for webpage
new_region_data = new_country_data.groupby('region').mean()
new_region_data
new_region_data = new_region_data.round(2)
new_region_data = new_region_data.reset_index()
new_region_data = new_region_data.drop(columns={'latitude','longitude','hdi_rank','happiness_rank_2015'})
```

```python
# export the dataframes to csvs
new_region_data.to_csv(f'data/new_region_data.csv', index=False)
new_country_data.to_csv(f'data/new_country_data.csv', index=False)
```
