```python
# Drop all the unneededs columns from all the dataframes
# alcohol = alcohol.drop(columns={'SUBJECT','MEASURE','FREQUENCY','Flag Codes','INDICATOR','TIME'})
doctors = doctors.drop(columns={'Indicator Name'})
health_spending = health_spending.drop(columns={'Indicator Name'})
# # social_support = social_support.drop(columns={'SUBJECT','MEASURE','FREQUENCY','Flag Codes','INDICATOR','TIME'})
# obese = obese.drop(columns={'SUBJECT','MEASURE','FREQUENCY','Flag Codes','INDICATOR','TIME'})
# years_lost = years_lost.drop(columns={'SUBJECT','MEASURE','FREQUENCY','Flag Codes','INDICATOR','TIME'})
# Rename the value column
alcohol = alcohol.rename(columns={'Entity':'country','Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)':'alcohol_consumption_percapita'})
doctors = doctors.rename(columns={'Country Name':'country','2015':'physicians_per_1k_2015','2016':'physicians_per_1k_2016','2017':'physicians_per_1k_2017'})
health_spending = health_spending.rename(columns={'Country Name':'country','2015':'health_spending_percapita_2015','2016':'health_spending_percapita_2016','2017':'health_spending_percapita_2017','2018':'health_spending_percapita_2018'})
debt = debt.rename(columns={'Household debt, loans and debt securities (Percent of GDP)':'country','2015':'household_debt_2015','2016':'household_debt_2016','2017':'household_debt_2017','2018':'household_debt_2018'})
# social_support = social_support.rename(columns={'Value':'social_support'})
# obese = obese.rename(columns={'Value':'obese'})
# years_lost = years_lost.rename(columns={'Value':'years_lost'})
```
